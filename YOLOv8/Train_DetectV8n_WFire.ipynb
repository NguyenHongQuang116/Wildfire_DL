```python
##===============================DETECT FROM live cam and save results as a video
from ultralytics import YOLO
import cv2
import time
import os

model = YOLO("C://Users//user//work//WILDFIRE//YOLOv8_detec//runs//detect//train6//weights//best.pt")
cap = cv2.VideoCapture(0)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID') # You can try other codecs if XVID doesn't work
out = cv2.VideoWriter('YOLOv8_detect_x1.avi', fourcc, 20.0, (640, 480)) # Adjust frame size as needed

start_time = time.time()
prediction_duration = 30

while (time.time() - start_time) < prediction_duration:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    plotted_img = results[0].plot()

    # Write the frame to the video file
    out.write(plotted_img)
    cv2.imshow('CCTV camera 0', plotted_img)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```
